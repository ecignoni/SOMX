```python
import os
os.environ['CUDA_VISIBLE_DEVICES'] = str(0)

%load_ext autoreload
%autoreload 2

import somx

import jax
from jax import random
import jax.numpy as jnp

from sklearn import datasets

import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
%matplotlib inline

from IPython.display import HTML
```

# Self-Organizing Map on the IRIS dataset

We use the first two dimensions

```python
iris = datasets.load_iris()
X = iris.data[:, :2]
y = iris.target
```

Dataset colored by labels

```python
fig, ax = plt.subplots(1, 1, figsize=(6, 3), dpi=150)
[b.set_linewidth(2) for b in ax.spines.values()]
plt.scatter(X[:,0], X[:,1], c=y, ec='k')
```

We use a 5 $\times$ 5 grid of prototypes. The learning rate is fixed, as well as the neighbor radius. We train for 100 rounds (100 passes through the whole dataset).

```python
prototypes_shape = jnp.array([5, 5])
learning_rate = 1e-1
neighbor_radius = 2
key = random.PRNGKey(0)
n_rounds = 100

som_res = somx.som_online(X,
                          prototypes_shape=prototypes_shape,
                          learning_rate=learning_rate,
                          neighbor_radius=neighbor_radius,
                          key=key,
                          n_rounds=n_rounds
                         )
```

The output of SOM gives us the topology of the SOM, the final prototypes, the history (values of the prototypes along the training), and the reconstruction error (along the training):

```python
som_top, som_prot, som_hist, som_recerr = som_res['topology'], som_res['prototypes'], som_res['history'], som_res['reconstruction_error']
```

Reconstruction error at the end of each epoch/round:

```python
fig, ax = plt.subplots(1, 1, figsize=(6, 3), dpi=150)
[b.set_linewidth(2) for b in ax.spines.values()]

ax.plot(som_recerr)
ax.set_xlabel('# Epoch')
ax.set_ylabel('Reconstruction error')
```

Topology of the SOM. Observations are assigned to the closest prototype. Observations are colored by their original label:

```python
somx.plot_som_space(som_top,
                    som_prot,
                    X,
                    key,
                    color_by=y
                   )
```

SOM prototypes locations in feature space:

```python
fig, ax = plt.subplots(1, 1, figsize=(6, 3), dpi=150)
[b.set_linewidth(2) for b in ax.spines.values()]

ax.scatter(X[:,0], X[:,1], c=y, label='Observations')
ax.scatter(som_prot[:,0], som_prot[:,1], c='red', ec='k', s=50, label='SOM prototypes')
ax.legend()
```

```python
def animate_scatter(arrays, frames=10, auxil_scatter=None, auxil_color=None, xlim=None, ylim=None):

    fig, ax = plt.subplots(1, 1, dpi=150)
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
    
    if auxil_scatter is not None:
        _ = ax.scatter(auxil_scatter[:,0], auxil_scatter[:,1], c=auxil_color)
    
    scatter = ax.scatter(arrays[0][:,0], arrays[0][:,1], c='r', ec='k')

    def animate(i):
        scatter.set_offsets(arrays[i])
        return scatter,
    
    anim = FuncAnimation(fig, animate, frames=frames)
    return anim
```

```python
anim = animate_scatter(som_hist, frames=100, auxil_scatter=X, auxil_color=y, xlim=(4, 8), ylim=(1.7, 4.7))
```
